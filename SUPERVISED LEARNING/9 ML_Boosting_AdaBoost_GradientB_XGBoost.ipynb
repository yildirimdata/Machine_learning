**1. Boosting Theory**

Boosting is a technique that can be applied to an existing machine learning algorithm to improve its performance and to create a strong learner. The most widely used algorithm with boosting is the decision tree, as it produces the best results when paired with boosting. **Boosting is an iterative method that builds a strong model by combining the outputs of weak learners, which are created on the same data set**. The purpose of boosting is **to improve the prediction power of the model by reducing the errors of the weak learners**.

Boosting works by **sequentially** training a series of weak learners on different subsets of the data. In other words, the weak learners are not trained independently of each other, but rather in a series where the next learner tries to correct the errors of the previous one. The final strong learner is a weighted combination of all the weak learners, where each learner's weight is determined by its accuracy. By combining the predictions of multiple weak learners, the strong learner can accurately classify data points even when the individual weak learners cannot.

Boosting is a popular technique in machine learning because it can improve the performance of many algorithms, including decision trees, neural networks, and support vector machines.

**Note:** In the **Bagging method**, trees obtain their results independently from each other. In **Boosting**, each tree uses the results of the previous one. **So trees are dependent on each other**.

One popular boosting algorithm is **Adaboost, which is an acronym for Adaptive Boosting**. It is a method that combines the output of weak learners such as decision trees to improve the performance of the model. **The basic idea of Adaboost is to adjust the weights of the training instances in each iteration, such that the weight of misclassified instances is increased and the weight of correctly classified instances is decreased**. This way, the algorithm focuses more on the hard-to-classify instances and improves its accuracy.

Another example of boosting algorithm is **Gradient Boosting (GBM) which uses gradient descent to minimize the loss function**, the algorithm creates a new tree that tries to correct the mistakes of the previous tree. It also use decision tree as a weak learner.

Overall, **Boosting is a powerful technique that can be applied to decision tree to improve its performance and make it a more accurate and robust algorithm. It gives more weight to the misclassified instances and focuses on correcting the mistakes of the previous models to improve the prediction power of the final model**.

# 2. AdaBoost (Adaptive Boosting) Theory

Adaboost, short for Adaptive Boosting, is a powerful ensemble learning algorithm that operates by creating an ensemble of weak learners and then combining them using a weighted sum. The algorithm uses a combination of boosting and decision trees to produce a more accurate and robust model.

**Adaboost works by iteratively adding weak learners (such as decision trees) to the ensemble, with each new learner focusing on the instances that were misclassified by the previous learner**. At each iteration, the algorithm assigns **higher weights to the misclassified instances, forcing the next learner to focus more on these instances**. **Weights can be determined using the error value. For example, the higher the error, the more weight assigned for the observation**. Total error will always be between 0, for a perfect stump, and 1, for a horrible stump. This process is continued until a stopping criterion (until the error function remains unchanged or the maximum limit of the number of estimators is reached)is met. **The final ensemble model is then a weighted sum of the individual weak learners**.

One of the key features of Adaboost is that **it uses a sequential approach to building the ensemble, rather than a parallel approach**. Rather than building all the trees at once and then combining them, Adaboost builds the trees one at a time, using the experience from the previous tree to guide the building of the next tree. This is what gives Adaboost its name, as it adapts to the data at each step.

As a result, Adaboost can be **more effective at reducing bias and variance** in the model by adapting to the data at each step. The adaptive nature of AdaBoost makes it particularly useful in scenarios where the data is complex and difficult to classify. Some main keywords are:

- Weak learner/estimator,
- Stump
- Ensemble

# 3) Gradient Boosting Theory

Gradient Boosting is an ensemble learning algorithm that is similar to Adaboost in that it **aims to correct the errors of the previous model by adding new models to the ensemble**. However, **unlike Adaboost, Gradient Boosting uses residual estimation errors to update the model. This means that instead of adjusting the weights of the previous model, Gradient Boosting uses the difference between the predicted value and the actual value to update the model**. (residual in lin reg vs pseudo residual in GB)

One important difference between Gradient Boosting and Adaboost is the way the algorithm builds the ensemble. In Gradient Boosting, the algorithm starts with a leaf, not a stump, and it allows for the creation of larger trees than Adaboost. Additionally, Gradient Boosting uses weak learners with the same learning rate, whereas Adaboost uses weak learners with different learning rates. 

Another key difference between the two algorithms is that the training for Gradient Boosting is based on the residuals of the previous model, whereas Adaboost trains using the misclassifications of the previous model. **The ensemble model is trained to minimize the sum of the squared errors** between its predictions and the actual values of the output variable.

Another thing is that **Gradient Boosting algorithm required more computational time as compared to Adaboost**, as the learning rate is typically set to a small value between 0 and 1. With a small learning rate, each subsequent tree added to the ensemble doesn't contribute much, and more trees are needed to achieve the same accuracy as Adaboost.

Some important differences of Gradient Boosting:

- Everything starts with a leaf, not a stump,
- We can create larger trees than Adaboost,
- Weak learners have the same learning rate,
- The training for gradual series learning is based on the residuals of the previous model.

Gradient boosting is a powerful machine learning technique that can be used for many different types of problems, including regression, classification, and ranking. It is particularly useful in scenarios where the data is complex and difficult to model, and where a single strong model cannot accurately capture the relationships between the input features and the output variable.

# 4) XGBoost Theory

**XGBoost (extreme Gradient Boosting) is an advanced implementation of the gradient boosting algorithm**. XGBoost has several improvements over traditional gradient boosting algorithms, making it a more powerful and efficient machine learning algorithm. It has become extremely popular in recent years, particularly in machine learning competitions, due to its high predictive power and fast computational speed. In simple words, **XGBoost is a powerful tool for creating accurate prediction models by combining many weak models into a strong ensemble model**.

XGBoost is designed to be scalable, efficient, and flexible, making it well-suited for a wide range of machine learning tasks. It supports **both regression and classification problems**, and can be used with a variety of different input data types, including numerical, categorical, and text data.

**One of the main advantages of XGBoost is its ability to handle large datasets with many features**. XGBoost is almost **10 times faster than other gradient boosting techniques**, making it an **ideal choice for large-scale datasets**. This is achieved **by using techniques such as parallel processing and caching**, which are not present in other gradient boosting algorithms.

Another feature that sets XGBoost apart from other gradient boosting algorithms is the way it handles overfitting. **XGBoost includes several adjustments that help to reduce overfitting and improve overall performance. For example, it uses regularization techniques such as L1 and L2 regularization to prevent overfitting. Additionally**, it uses a technique called **'stochastic gradient boosting'** which utilizes a randomly selected subset of the training data in each iteration to reduce overfitting. In traditional boosting, each new model in the sequence is trained on the entire training set, which can lead to overfitting if the model is too complex or if there are noisy or irrelevant features in the data. In Stochastic Gradient Boosting, a random subset of the training data is sampled for each new model in the sequence. This helps to reduce overfitting and can improve the generalization performance of the model.

**XGBoost also implements a technique called "Out-of-Bag" evaluation**, which is a way to estimate the generalization performance of the model by evaluating the predictions of each instance against the instances that were not used in the training of that tree.

Overall, XGBoost is a powerful and efficient machine learning algorithm that has proven to be extremely effective in machine learning competitions and industry setting. **With its high predictive power, computational efficiency and adjustment for overfitting, it is often considered as the "regular boost" technique.**

How XGBoost is comparatively better than other techniques:

- **Regularization:**  XGBoost includes several types of regularization to prevent overfitting and improve the generalization ability of the model. The standard GBM implementation has no regulation like XGBoost.

- **Parallel Processing:** XGBoost implements parallel processing and is faster than GBM. XGBoost can be run in parallel on multiple CPUs, which makes it faster and more scalable than some other machine learning libraries.

- **High Flexibility:** XGBoost allows users to define custom optimization targets and evaluation criteria that add a new dimension to the model.

- **Handling Missing Values:** XGBoost has a built-in routine to handle missing values. we don't need to fill in the missing values, as XGBoost takes care of them.

- **Tree Pruning:** XGBoost creates splits up to the specified max_depth and then starts pruning back the tree, removing splits beyond which there is no positive gain.

    - Max_depth = Used to define the maximum depth. Higher depth will allow the model to learn relationships that are specific to a particular instance.

- **Built-in Cross-Validation:** XGBoost allows a user to cross-validate each iteration of the upgrade process so it's easy to get exactly the optimum number of upgrade iterations in a single transaction.

- **Interpretability:** XGBoost provides a number of tools for understanding how the model is making predictions, including feature importance scores and visualization tools.

XGBoost has been used successfully in a wide range of applications, including predictive maintenance, fraud detection, and natural language processing. It has also won numerous machine learning competitions on Kaggle, demonstrating its effectiveness and versatility.

**Key Terms**

**1) Weak learner**

A weak learner is a simple machine learning algorithm that performs slightly better than random guessing on a given task. In simple words, a weak learner is a basic model that can make predictions that are better than random chance, but may not be accurate enough to use on its own for real-world applications.

Examples of weak learners include decision trees with small depths, simple linear regression models, or random classifiers. These models are often used as building blocks in ensemble learning methods, such as AdaBoost and Random Forests, which combine multiple weak learners to create a stronger, more accurate model.

The idea behind using weak learners in ensemble methods is that by combining many weak learners, each of which has a small predictive power, the ensemble model can achieve a higher level of accuracy than any single weak learner. Weak learners are particularly useful in scenarios where the data is noisy, complex, or high-dimensional, and when it's hard to identify a single strong model that can accurately classify the data.

**2) Stump**

A stump, refers to a decision tree with a single split.A stump, being a decision tree with only one split, is limited in its ability to model complex relationships between the input features and the output variable. However, stumps can be useful building blocks in ensemble methods like AdaBoost, where many stumps are combined to create a stronger model.

**3) Ensemble**

An ensemble, in the context of machine learning, refers to a technique of combining multiple models to create a more accurate and robust prediction model. In simple words, an ensemble is a group of models that work together to make predictions, with the goal of achieving higher accuracy than any single model could achieve alone.

Ensemble methods often involve combining multiple models of the same type, such as decision trees or neural networks, or using different types of models, such as combining decision trees with logistic regression models. The idea behind ensemble methods is that by combining the predictions of many individual models, the overall prediction is less prone to error and more robust to fluctuations in the data.

Ensemble methods can be divided into two categories: bagging and boosting. Bagging methods involve training multiple models independently on different subsets of the data and then combining their predictions through voting or averaging. Boosting methods, on the other hand, train multiple weak models sequentially and adaptively, with each subsequent model focusing on the errors of the previous models.

Ensemble methods have been successfully applied to many different types of machine learning problems, including classification, regression, and anomaly detection. They are particularly useful in scenarios where the data is noisy or complex, and when it's hard to identify a single strong model that can accurately classify or predict the data.

# Similarities and differences between bagging and boosting

**Bagging and boosting** are two popular ensemble methods in machine learning used to improve the performance of predictive models. While both techniques are designed to improve the accuracy of the model, they do so in different ways. Here are some similarities and differences between bagging and boosting:

**Similarities:**

- Both bagging and boosting are ensemble methods that combine multiple models to improve performance.
- They both involve building multiple models on subsets of the training data and aggregating their predictions to produce a final prediction.
- Both techniques are effective at reducing overfitting.

**Differences:**

- Bagging and boosting differ in the way they create and combine models. Bagging builds multiple models independently, each on a randomly sampled subset of the training data. Boosting builds models sequentially, with each model attempting to correct the errors of the previous models.
- Bagging reduces overfitting by averaging the predictions of multiple models, while boosting reduces bias by focusing on the examples that are difficult to predict correctly.
- Bagging is more effective for unstable models that are sensitive to changes in the training data, while boosting is more effective for stable models that consistently perform well on the training data.
- Bagging can be parallelized, making it faster to train models on large datasets, while boosting cannot be parallelized and requires more computation time.
- Bagging tends to have lower variance but higher bias, while boosting tends to have higher variance but lower bias.

In summary, bagging and boosting are both effective techniques for improving the performance of predictive models, but they differ in the way they create and combine models, their strengths and weaknesses, and the types of models they are best suited for.

```python
y = df['DEFAULT']
X = df.drop(['DEFAULT','ID'], axis=1) # becareful inplace= False

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=rand_state)
```

# Training the models

```python
%%time 
# Fitting RF classifier to the Training set
RF_classifier = RandomForestClassifier(random_state=rand_state)
RF_classifier.fit(X_train, y_train)
```

```python
%%time
# Fitting AdaBoost classifier to the Training set
AdB_classifier = AdaBoostClassifier(random_state=rand_state)
AdB_classifier.fit(X_train, y_train)
```

```python
%%time
# Fitting Gradient Boosting classifier to the Training set
GBM_classifier = GradientBoostingClassifier(random_state=rand_state, learning_rate=0.1, max_depth=6)
GBM_classifier.fit(X_train, y_train)
```

```python
%%time
# Fitting XGBoost classifier to the Training set
XGB_classifier = XGBClassifier(random_state=rand_state, eta=0.1, max_depth=6)
XGB_classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set probabilities and classes
y_hat_RF       = RF_classifier.predict(X_test)
y_hat_AdB      = AdB_classifier.predict(X_test)
y_hat_GBM      = GBM_classifier.predict(X_test)
y_hat_XGB      = XGB_classifier.predict(X_test)
```

# Evaluating the model performance on test data

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
```

```python
print('RF  Accuracy = {}'.format(accuracy_score(y_test, y_hat_RF)))
print('AdB Accuracy = {}'.format(accuracy_score(y_test, y_hat_AdB)))
print('GBM Accuracy = {}'.format(accuracy_score(y_test, y_hat_GBM)))
print('XGB Accuracy = {}'.format(accuracy_score(y_test, y_hat_XGB)))
```

```python
print('RF  f1 = {}'.format(f1_score(y_test, y_hat_RF)))
print('AdB f1 = {}'.format(f1_score(y_test, y_hat_AdB)))
print('GBM f1 = {}'.format(f1_score(y_test, y_hat_GBM)))
print('XGB f1 = {}'.format(f1_score(y_test, y_hat_XGB)))
```
