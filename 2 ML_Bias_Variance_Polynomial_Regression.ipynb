# Bias-Variance Trade-Off - 1

The bias-variance tradeoff is a fundamental principle in supervised machine learning that describes **the balance between a model's ability to fit the training data accurately and its ability to generalize to new, unseen data**. High bias models have a strong tendency to underfit the data, meaning they cannot capture the complexity of the underlying relationship in the data and perform poorly on both the training and unseen data. High variance models, on the other hand, have a strong tendency to overfit the data, meaning they fit the noise in the training data and perform well on the training data but poorly on unseen data.

Bias, modelin gerek deerden ne kadar sapma gsterdiini ifade eder. Yksek bias, modelin ar basit olduu ve verilerin karmakln yeterince yanstamad anlamna gelir. Buna karlk, dk bias, modelin verileri iyi yanstt anlamna gelir.

A well-balanced model, which is the goal of machine learning, will have the right amount of bias and variance that allows the model to learn the underlying structure of the data, allowing the model to make accurate predictions on new unseen data.

Functional structures or curves, created based on the features, give insights into the performance of the model. The purpose of creating a successful and balanced model is to enable our model to learn the general structure of the data. Learning the general structure of the data is critical, as it enables the model to make predictions on new data with minimal error. **Underfitting and Overfitting Problems**

**High Bias and Underfitting Problem**

In simple terms, **bias is the difference between the predicted values of a model and the true values of the data**. When examining the second graph, it is clear that the distances between many of the real values and the estimated values (represented by the red curve) are quite large. Additionally, it can be seen that the red curve does not conform to the normal distribution of the data and has a high bias towards certain observations.

This suggests that the curve generated by our model does not accurately represent the underlying data, and as a result, our model has not learned the underlying structure of the data sufficiently. This is referred to as underfitting, where a model is not complex enough to capture the complexity of the data. 

**Tips-1:** In underfitting, the model is not able to learn the data adequately and therefore cannot represent the structure of the data sufficiently.

**High Variance and Overfitting Problem:**

**Variance refers to the flexibility and variability of a model in fitting the data. High variance models are highly flexible and can closely fit the training data**, as can be seen in the first graph. Though it may seem like a desirable trait, it causes the model to over-learn or memorize the training data, as it becomes too closely tied to the specific input data.

Overfitting: modelin eitim verilerine ar uyduu ve farkl veri kmesinde kt performans gsterebilecei anlamna geliyor.

PS: when we talk about variance, we always talk about test error. So if 2 randomly chosen model has test errors of 100 and 27, that means there is high variance due to overfitting. When we talk about bias, we always talk about train error. If train error gets bigger, bias gets bigger too. Balanced fitting means low bias and low variance. 

A well-performing model should be able to learn the underlying structure of the data, not just memorize it. When a model is overfitting, it becomes too specific to the training data, and would not be able to generalize to new, unseen data. As a result, it would produce inaccurate results when applied to new data, as it was not designed to handle it.        

**Tips-2:** In Overfitting, the model learns the data in all its details, in other words, it memorizes the data. It's not a good thing, because, in this way, the model is not a general model that represents the data structure, just a special model of that specific data.

**Can a machine learning model have low bias and low variance?**

In general, it is not possible for a machine learning model to have both low bias and low variance, because the bias-variance tradeoff is a fundamental principle of supervised learning. However, **by carefully balancing the model complexity and the amount of regularization, it is possible to find a model that has a good balance of bias and variance, and therefore achieves good performance on the training data and generalizes well to new, unseen data**.



**Example: Underfitting and Overfitting Students**

Imagine two students taking an exam, the first student has not studied the subject sufficiently and therefore is unable to provide clear and correct answers to the exam questions, this student is underfitting the exam. He was not able to fully grasp and understand the subject.

On the other hand, the second student has memorized all past exam questions but has not truly understood the subject, so when faced with a new question on the exam, they will not be able to correctly answer it as they have not learned the general structure of the subject. They just memorized the previous questions, and this is called overfitting.

**Training Error vs. Validation Error (Test Error)**

In the Linear Regression lesson, we learned about the concept of train and test errors, which are important measures of a model's performance. These errors are closely related to the concepts of overfitting and underfitting and play a crucial role in building a well-balanced and successful model.



One of the most important goals in building a machine learning model is to minimize the error rate, so that the model can make accurate predictions. To achieve this, we can make changes to the model's parameters and features, which will ultimately improve its performance. However, as we make these changes, the model becomes more complex, increasing the risk of overfitting, which is a high variance model. **This trade-off between reducing errors and increasing model complexity is known as the bias-variance trade-off.**

As shown in the picture above, there is a balance area between high bias and high variance, where the training and test errors are in balance. # Polynomial Regression

Polynomial Regression is a type of regression analysis in which the relationship between the independent variable x and the dependent variable y is modeled as an nth degree polynomial. It is particularly useful when the **relationship between the variables is non-linear**.



Unlike linear regression models, polynomial regression models are able to capture complex non-linear relationships in the data by artificially creating new features through mathematical transformations. These new features allow the model to fit a non-linear function to the data.

In polynomial regression, **the degree of the polynomial is a crucial factor** to consider. The higher the degree, the more complex the model, and it can also lead to overfitting, which makes the model overly specific to the training data, but performs poorly on unseen data. On the other hand, when the degree is set to 1, it results in a simple linear regression model.

**Polynomial Regression**

Polynomial regression extends the linear model by adding extra predictors, obtainded by raising each of the original predictors to a power. This approach provides a simple way to provide a non-linear fit to data. The model however is still linear.

**All in One Function**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split


def my_polynomial_regression(poly_degree, X, y, test_size=0.3, rand_state=100):
    
    poly_features= PolynomialFeatures(degree=poly_degree)
    
    # transforms the existing features to higher degree features.
    X_poly = pd.DataFrame(poly_features.fit_transform(X.reshape(-1,1)))   
    
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=test_size, random_state=rand_state)

    # fit the transformed features to Linear Regression
    poly_model = LinearRegression()
    poly_model.fit(X_train, y_train)
    
  
    # predicting on training data-set
    y_train_predicted = poly_model.predict(X_train)
  
    # predicting on test data-set
    y_test_predicted = poly_model.predict(X_test)
  
   # evaluating the model on training dataset
    MSE_train = round(np.mean(np.square(y_train - y_train_predicted)),2)
    RMSE_train = round(np.sqrt(MSE_train),2)

  
    # evaluating the model on test dataset
    MSE_test = round(np.mean(np.square(y_test - y_test_predicted)),2)
    RMSE_test = round(np.sqrt(MSE_test),2)


  
#     print("Traning model fit")
#     print("R2_train = {}".format(R2_train))

#     print("-------------------------------------------")
#     print("MSE_train = {}".format(MSE_train))
#     print("RMSE_train = {}".format(RMSE_train))

#     print("-------------------------------------------")
#     print("MSE_test = {}".format(MSE_test))
#     print("RMSE_test = {}".format(RMSE_test))
    
    return poly_degree, RMSE_train, RMSE_test
```

```python
my_polynomial_regression(poly_degree=5,X=x,y=y)
```

```python
my_polynomial_regression(poly_degree=1, X=x, y=y)
```

```python
returns= []
for i in range(1,6):
    returns.append(my_polynomial_regression(poly_degree=i, X=x,y=y))
    
output = pd.DataFrame(returns, columns=['Degree', 'RMSE_train', 'RMSE_test'])
output

# RMSE_test is important for us. As it can be seen the 3rd polynomial degree is the optimal point. Afterthat it increases
```

```python
output.set_index('Degree').plot()
plt.show()
```

**Cross validation**

```python
from sklearn.model_selection import cross_val_score
import sklearn.metrics
```

```python
# recall our d=5 so far! 
X_train.head()
```

```python
NMSE = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 10 , scoring="neg_mean_squared_error" )
# cv is the number of Folds. To speed up the process you can set the n_jobs=-1
# to see the list of valid scorings try:sorted(sklearn.metrics.SCORERS.keys())
```

```python
NMSE
```

```python
# mse cross validation
MSE_CV= np.mean(-NMSE)
MSE_CV
```

```python
RMSE_CV = np.sqrt(-NMSE).mean()
RMSE_CV
```
